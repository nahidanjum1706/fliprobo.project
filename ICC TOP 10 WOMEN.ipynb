ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
players=[]
teams=[]
ratings=[]
```

```python
#Extract the tags of player for 1st row

player_name = soup.find_all('div',class_='rankings-block__banner--name-large')
for i in player_name:
    players.append(i.text.strip())
```

```python
#Extract the tags of remaining players from 2nd row

player_name = soup.find_all('td',class_='table-body__cell rankings-table__name name')
for i in player_name:
    players.append(i.text.strip())
    players = players[:10]
```

```python
#Extract the tags of Team name for 1st row

team_names = soup.find_all('div',class_='rankings-block__banner--nationality')
for i in team_names:
    teams.append(i.text.strip())
```

```python
#Extract the tags of remaining Team names from 2nd row

team_names = soup.find_all('span',class_='table-body__logo-text')
for i in team_names:
    teams.append(i.text.strip())
    teams = teams[:10]
```

```python
#Extract the tags of rating for 1st row

rating = soup.find_all('div',class_='rankings-block__banner--rating')
for i in rating:
    ratings.append(i.text.strip())
```

```python
#Extract the tags of remaining ratings from 2nd row

rating = soup.find_all('td',class_='table-body__cell rating')
for i in rating:
    ratings.append(i.text.strip())
    ratings = ratings[:10]
```

```python
Top_Batsmen_ODI = pd.DataFrame({})
Top_Batsmen_ODI['players'] = players
Top_Batsmen_ODI['teams'] = teams
Top_Batsmen_ODI['ratings'] = ratings
Top_Batsmen_ODI
```

iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/bowling")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
players=[]
Teams=[]
ratings=[]
```

```python
#Extract the tags of player name for 1st row

player=soup.find_all('div',class_='rankings-block__banner--name-large')
for i in player:
    players.append(i.text)
    
#Extract the tags of remaining player name from 2nd row

player=soup.find_all('td',class_='table-body__cell rankings-table__name name')
for i in player:
    players.append(i.find('a').text)
    players=players[:10]
```

```python
#Extract the tags of team name for 1st row

team = soup.find_all('div',class_='rankings-block__banner--nationality')
for i in team:
    Teams.append(i.text.replace('\n','').strip())
    
#Extract the tags of remaining team name from 2nd row

team = soup.find_all('span',class_='table-body__logo-text')
for i in team:
    Teams.append(i.text.replace('\n','').strip())
    Teams = Teams[:10]
```

```python
#Extract the tags of rating for 1st row

rating=soup.find_all('div',class_='rankings-block__banner--rating')
for i in rating:
    ratings.append(i.text)
    
#Extract the tags of remaining ratings from 2nd row

rating=soup.find_all('td',class_='table-body__cell rating')
for i in rating:
    ratings.append(i.text)
    ratings=ratings[:10]
```

```python
Bowlers_Top_ODI=pd.DataFrame({})
Bowlers_Top_ODI['players']=players
Bowlers_Top_ODI['Teams']=Teams
Bowlers_Top_ODI['ratings']=ratings
Bowlers_Top_ODI
```
